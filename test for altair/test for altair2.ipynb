```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
np.random.seed(10)
value = np.random.randn(365)
data = np.cumsum(value)
date = pd.date_range(start="20190101", end="20191231")
df = pd.DataFrame({"cumsum":data}, index=date)
```

```python
chart = alt.Chart(df.reset_index(), height=300, width=500).mark_line().encode(
    x="index:T", y="cumsum:Q"
)
chart
```

```python
chart = alt.Chart(df.reset_index(), height=300, width=500).mark_line().encode(
    x = alt.X("index:T", axis=alt.Axis(format="%b"), title="month"),
    y = alt.Y("cumsum:Q", aggregate="max", title="Monthly Max cumSum", type='quantitative')
)
chart
```

```python
chart = alt.Chart(df.reset_index(), height=300, width=500).mark_bar().encode(
    x = alt.X("month(index):O", axis=alt.Axis(format="%b", formatType="time"), title="month"),
    y = alt.Y("maxCumSum:Q", title="Monthly Max CumulativeSum"),
    color = alt.condition(alt.datum.maxCumSum>0,
                          alt.value("steelblue"), alt.value("orange"))
).transform_timeunit(
    month="month(index)"
).transform_aggregate(
    maxCumSum = "max(cumsum)", groupby=["month"]
)
chart
```
