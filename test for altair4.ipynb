```python
import pandas as pd
import altair as alt

data = {"Year":[2016, 2017],
        "Region A":[2.5, 2.1],
        "Region B":[2.4, 2.5],
        "Region C":[0.7, 1.1],
        "Region D":[2.1, 2.7],
        "Region E":[0.9, 1.4],
        "Region F":[3.4, 3.3],
        }

df = pd.DataFrame(data, index=None)
df = df.melt("Year", var_name="Region", value_name="Sales")
df.head()
```

```python
chart = alt.Chart(df, width=200).mark_line(point=True).encode(
    alt.X("Year:O", axis=alt.Axis(labelAngle=15)),
    y = "Sales:Q",
    color = "Region:N"
)

chart
```

```python
project = [
    {"project": "1", "start":"2018-01-18", "end":"2018-03-08"},
    {"project": "2", "start":"2017-07-28", "end":"2019-05-05"},
    {"project": "3", "start":"2016-02-02", "end":"2020-09-15"},
    {"project": "4", "start":"2013-01-22", "end":"2016-12-19"},
    {"project": "5", "start":"2017-01-16", "end":"2022-04-16"},
    {"project": "6", "start":"2011-01-08", "end":"2011-06-22"},
    {"project": "6", "start":"2018-01-08", "end":"2019-07-02"},
]
df = alt.Data(values=project)
```

```python
df
```

```python
chart = alt.Chart(df, width=500, height=150).mark_bar().encode(
    alt.X("start:T", axis=alt.Axis(format="%x", formatType="time",
                                    tickCount=10, title=None),
            scale=alt.Scale(
                domain=[alt.DateTime(year=2011, month=1, date=1),
                        alt.DateTime(year=2022, month=5, date=31)]
            )),
    alt.X2("end:T"),
    alt.Y("project:N", axis=alt.Axis(labelAlign="left", labelFont="Bodoni MT",
                                     labelFontSize=15, labelOffset=0,
                                     labelPadding=112, title=None)),
    color=alt.Color("project:N", legend=alt.Legend(labelFont="Bodoni MT",
                                                   labelFontSize=12,
                                                   symbolOpacity=0.7,
                                                   titleFont="Bodoni MT",
                                                   titleFontSize=15))
)

chart
```
